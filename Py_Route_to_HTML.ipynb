### Visualize & Save a route (GPX) with Folium and OpenStreetMaps

```python
# with open('HU-GR_gpx/HU-GR.gpx', 'r') as gpx_file:
#     gpx = gpxpy.parse(gpx_file)

with open('My_Routes/PL-SK-HU.gpx', 'r') as gpx_file:
    gpx = gpxpy.parse(gpx_file)
```

```python
route_info = []

for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            route_info.append({
                'latitude': point.latitude,
                'longitude': point.longitude,
                'elevation': point.elevation
            })

route_df = pd.DataFrame(route_info)
#route_df.head()            
```

```python
route_map = folium.Map(
    location=[47, 32],
    zoom_start=5,
    tiles='OpenStreetMap',
    width=924,
    height=600
)

coordinates = [tuple(x) for x in route_df[['latitude', 'longitude']].to_numpy()]
folium.PolyLine(coordinates, weight=3).add_to(route_map)
folium.LayerControl().add_to(route_map)

display(route_map)

#route_map.save('./output_map.html')
```

### Routes (+1) plots with Folium and OpenStreetMaps

#### Load the different routes

```python
with open('My_Routes/PL-SK-HU.gpx', 'r') as gpx_file:
    gpx = gpxpy.parse(gpx_file)

route_info = []

for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            route_info.append({
                'latitude': point.latitude,
                'longitude': point.longitude,
                'elevation': point.elevation
            })

route_df_y22 = pd.DataFrame(route_info)

route_df_y22['Performed'] = 'Y22'


route_df_y22.head(2)                
```

```python
with open('HU-GR_gpx/HU-GR.gpx', 'r') as gpx_file:
    gpx = gpxpy.parse(gpx_file)

route_info = []

for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            route_info.append({
                'latitude': point.latitude,
                'longitude': point.longitude,
                'elevation': point.elevation
            })

route_df_y24 = pd.DataFrame(route_info)

route_df_y24['Performed'] = 'Y24'

route_df_y24.head(2)                
```

```python
route_map = folium.Map(
    location=[47, 32],
    zoom_start=5,
    tiles='OpenStreetMap',
    width=924,
    height=600
)

coordinates = [tuple(x) for x in route_df_y22[['latitude', 'longitude']].to_numpy()]
folium.PolyLine(coordinates, weight=3,color='green').add_to(route_map)
folium.LayerControl().add_to(route_map)

coordinates = [tuple(x) for x in route_df_y24[['latitude', 'longitude']].to_numpy()]
folium.PolyLine(coordinates, weight=3, color='gray').add_to(route_map)
folium.LayerControl().add_to(route_map)

display(route_map)

#route_map.save('./output_map.html')
```
