```python
import subprocess
import json

def get_gps_speed(video_file):
    exiftool_path = "/usr/bin/exiftool"  # Or the full path to your exiftool executable
    command = [
        exiftool_path,
        "-ee",
        "-G1",
        "-json",
        video_file
    ]

    try:
        result = subprocess.run(command, capture_output=True, text=True, check=True)
        metadata_list = json.loads(result.stdout)  # Load as a list of metadata objects

        gps_speeds = []

        for metadata in metadata_list:
            if "Track4" in metadata and isinstance(metadata["Track4"], dict):
                track4_data = metadata["Track4"]

                # Check for the 'GoPro MET' handler description
                if "HandlerDescription" in track4_data and track4_data["HandlerDescription"] == "GoPro MET":
                    # Extract GPSSpeed if present
                    if "GPSSpeed" in track4_data:
                        gps_speed = track4_data["GPSSpeed"]
                        gps_speeds.append(gps_speed)

        return gps_speeds

    except subprocess.CalledProcessError as e:
        print(f"Error executing ExifTool: {e}")
        return []
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        return []
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return []

# Example usage
video_file = "/home/jalcocert/Desktop/Karting/GX030390.MP4"
speeds = get_gps_speed(video_file)

if speeds:
    print("GPS Speeds:", speeds)
else:
    print("No GPS speed data found or an error occurred.")
```

```python
import re

def extract_gps_speeds(file_path):
    gps_speeds = []
    try:
        with open(file_path, 'r') as file:
            for line in file:
                if "GPS Speed" in line:
                    parts = line.strip().split(':', 1)  # Split on the first colon
                    if len(parts) == 2:
                        speed_str = parts[1].strip()

                        # Use regex to find numbers, ignoring any non-numeric characters (like units)
                        match = re.findall(r"[\d\.]+", speed_str)
                        if match:
                            try:
                                speed = float(match[0])  # Convert the first match to a float
                                gps_speeds.append(speed)
                            except ValueError:
                                print(f"Error parsing speed from: {line.strip()}")
                        else:
                            print(f"Could not extract a numeric value from: {line.strip()}")

        print("Extracted GPS Speeds:")
        for speed in gps_speeds:
            print(speed)

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

file_path = "./Data_GoPro/output-GX030390.txt"
extract_gps_speeds(file_path)
```

```python
import plotly.express as px

# Extract GPS speeds
gps_speeds = extract_gps_speeds(file_path)

# If there are speeds to plot, create a boxplot
if gps_speeds:
    fig = px.box(y=gps_speeds, labels={'y': 'GPS Speeds (km/h)'}, title="Boxplot of GPS Speeds")
    fig.show()
else:
    print("No GPS speeds found to plot.")
```
