```python
#https://stackoverflow.com/questions/72130267/how-to-modify-points-drawn-on-map-using-a-dropdown-menu

import numpy as np
import pandas as pd
import plotly.express as px #does not support drop-downs
import plotly.graph_objects as go #we will use this one for drop-downs

us_cities = pd.read_csv(
    'https://raw.githubusercontent.com/plotly/datasets/master/us-cities-top-1k.csv'
)

traces = []
buttons = []
state_list = np.append('All_state', us_cities['State'].unique())
visible = state_list

for s in state_list:
    #print(s)
    if s == 'All_state':
        filtered_df = us_cities.copy()
    else:
        filtered_df = us_cities[us_cities['State'] == s]

    traces.append(go.Scattermapbox(
        lat=filtered_df['lat'],
        lon=filtered_df['lon'],
        mode='markers',
        visible=True if s == state_list[0] else False,
        customdata=filtered_df,
        hovertemplate='City: %{customdata[0]}<br>Population: %{customdata[2]}<extra></extra>',
        marker=go.scattermapbox.Marker(
            size=9,
            color='fuchsia'
        )
    ))
    
    buttons.append(
        dict(
            method='update',
            label=s,
            args=[{'visible':list(visible==s)}],)
    )

fig = go.Figure(data=traces)
fig.update_layout(
    mapbox=dict(
        style='open-street-map',
        #accesstoken=mapbox_access_token,
        bearing=0,
        center=go.layout.mapbox.Center(
            lat=us_cities['lat'].mean(),
            lon=us_cities['lon'].mean(),
        ),
        zoom=3
    ),
    margin={'r':10, 't': 0, 'l': 0, 'b': 0}
)

fig.update_layout(
    # autosize=False,
    height=1000,
    showlegend=False,
    updatemenus=[
        dict(
            buttons=buttons,
            direction='down',
            x=0.05,
            y=1.0,
            xanchor='right',
            yanchor='bottom',
            font=dict(size=8)
        ),
    ]
)

fig.show()
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px



#Creating Dash Application
app = JupyterDash(__name__)
#app = dash.Dash()


app.layout = html.Div([
    html.H4('Interactive scatter plot with Iris dataset'),
    dcc.Graph(id="scatter-plot"),
    html.P("Filter by petal width:"),
    dcc.RangeSlider(
        id='range-slider',
        min=0, max=2.5, step=0.1,
        marks={0: '0', 2.5: '2.5'},
        value=[0.5, 2]
    ),
])


@app.callback(
    Output("scatter-plot", "figure"), 
    Input("range-slider", "value"))

def update_bar_chart(slider_range):
    df = px.data.iris() # replace with your own data source
    low, high = slider_range
    
    mask = (df['petal_width'] > low) & (df['petal_width'] < high)

    fig = px.scatter(
        df[mask], x="sepal_width", y="sepal_length", 
        color="species", size='petal_length', 
        hover_data=['petal_width'])
    
    return fig



#Dash default port is 8050 ===> 127.0.0.1:8050

# Start of the application
if __name__ == '__main__':
    #app.run_server(debug=False)
    app.run_server(mode="inline")
    #app.run_server(mode="jupyterlab")      
```

#### Bar plot with DASH and Filters

```python
df = dataset[["latitude", "longitude", "price","room_type","neighbourhood_group"]].head(500)


#distinct values of year
neig_available = df['neighbourhood_group'].unique()



#Declaring and including the stylesheets
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

#Creating Dash Application
app = JupyterDash(__name__, external_stylesheets=external_stylesheets)
#app = dash.Dash(external_stylesheets=external_stylesheets)


colors = {
    'background': '#d48826',
    'text': '#7FDBFF'
}

#Creating Dash Application Layout
app.layout = html.Div(style={'backgroundColor': colors['background']},children=[
       html.H1('Welcome to Dash!'),
    
       html.P('Choose a Neighbourhood:'),
       dcc.Dropdown(id = 'dropdown-cell', 
                    options = [{'label': i, 'value': i} for i in neig_available],
                               multi = False, 
                               value = neig_available[0]),
       html.Br(),
       dcc.Graph(id = 'interactive-graph'),
       #dcc.Graph(figure=fig)
])

#Callback functions for interactivity between components
@app.callback(
    Output(component_id='interactive-graph', component_property='figure'),
    [Input(component_id='dropdown-cell', component_property='value')]
)

#Function which gets triggered automatically in case input box value is changed
# Here the xaxis contains the period corresponding to selected Cell ID
# and yaxis contains the XZ-Distance for the period for that selected Cell ID

def update_bar_graph(dropdown_value):
    
    # Get the data corresponding to the selection in the dropdown
    #roomtype_list = list(df.loc[df['neighbourhood_group'] == dropdown_value, "room_type"])
    #price_list = list(df.loc[df['neighbourhood_group'] == dropdown_value, "price"])

    df_grouped = df.loc[df['neighbourhood_group'] == dropdown_value]\
                 .groupby(['neighbourhood_group','room_type'], as_index=False).agg({'price': ['mean', 'min', 'max']})
    #df_grouped

    df_grouped.columns = [' '.join(col).strip() for col in df_grouped.columns.values]

    # roomtype_list = list(df_grouped.loc[df_grouped['neighbourhood_group'] == dropdown_value, "room_type"])
    # price_list = list(df_grouped.groupby(['neighbourhood_group','room_type']).mean()['price'])

    roomtype_list = list(df_grouped["room_type"])
    price_list = list(df_grouped["price mean"])


    # Figure information for the graph
    figure = {
        'data' : [
                    {'x' : roomtype_list , 'y' : price_list,
                    'type' : 'bar',
                    'textposition' : 'bottom',
                    'name' : dropdown_value
                    }
                ],
        'layout' : { 'showlegend' : True,
                   'xaxis' : {'title' : 'Room Type', 'linecolor' : 'grey', 'linewidth': 2, 'mirror' : True },
                   'yaxis' : {'title' : 'Price' ,'linecolor' : 'grey', 'linewidth': 2, 'mirror' : True },
                   'title' : 'Interactive Scatter Plot',
                   'height' : 500 
                   }
            }
    return figure


#Dash default port is 8050 ===> 127.0.0.1:8050

# Start of the application
if __name__ == '__main__':
    #app.run_server(debug=False)
    app.run_server(mode="inline")
    #app.run_server(mode="jupyterlab")        
```

```python
#https://stackoverflow.com/questions/72130267/how-to-modify-points-drawn-on-map-using-a-dropdown-menu

import numpy as np
import pandas as pd
import plotly.express as px #does not support drop-downs
import plotly.graph_objects as go #we will use this one for drop-downs

us_cities = pd.read_csv(
    'https://raw.githubusercontent.com/plotly/datasets/master/us-cities-top-1k.csv'
)

traces = []
buttons = []
state_list = np.append('All_state', us_cities['State'].unique())
visible = state_list

for s in state_list:
    #print(s)
    if s == 'All_state':
        filtered_df = us_cities.copy()
    else:
        filtered_df = us_cities[us_cities['State'] == s]

    traces.append(go.Scattermapbox(
        lat=filtered_df['lat'],
        lon=filtered_df['lon'],
        mode='markers',
        visible=True if s == state_list[0] else False,
        customdata=filtered_df,
        hovertemplate='City: %{customdata[0]}<br>Population: %{customdata[2]}<extra></extra>',
        marker=go.scattermapbox.Marker(
            size=9,
            color='fuchsia'
        )
    ))
    
    buttons.append(
        dict(
            method='update',
            label=s,
            args=[{'visible':list(visible==s)}],)
    )

fig = go.Figure(data=traces)
fig.update_layout(
    mapbox=dict(
        style='open-street-map',
        #accesstoken=mapbox_access_token,
        bearing=0,
        center=go.layout.mapbox.Center(
            lat=us_cities['lat'].mean(),
            lon=us_cities['lon'].mean(),
        ),
        zoom=3
    ),
    margin={'r':10, 't': 0, 'l': 0, 'b': 0}
)

fig.update_layout(
    # autosize=False,
    height=1000,
    showlegend=False,
    updatemenus=[
        dict(
            buttons=buttons,
            direction='down',
            x=0.05,
            y=1.0,
            xanchor='right',
            yanchor='bottom',
            font=dict(size=8)
        ),
    ]
)

fig.show()
```
