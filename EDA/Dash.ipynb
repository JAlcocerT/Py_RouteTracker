```python
import pandas as pd
import plotly.express as px
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import folium
from folium.plugins import MarkerCluster

from jupyter_dash import JupyterDash

# Define the dataframe
df = pd.DataFrame({
    "City": ["New York", "Paris", "London", "Tokyo", "São Paulo"],
    "Latitude": [40.7128, 48.8566, 51.5072, 35.6762, -23.5505],
    "Longitude": [-74.0060, 2.3522, -0.1278, 139.6503, -46.6333],
    "GDP per capita (2020)": [75200, 54500, 56500, 42700, 16200]
})

external_stylesheets = ['https://stackpath.bootstrapcdn.com/bootstrap/4.5.2/css/bootstrap.min.css']
#app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

# Create app
# app = dash.Dash()
# Build App
app = JupyterDash(__name__, external_stylesheets=external_stylesheets)

# Define the layout
app.layout = html.Div([
    html.H1("World Cities Explorer"),
    html.Div([
        html.Label("Select cities to display on map:"),
        dcc.Dropdown(
            id="city-dropdown",
            options=[{"label": city, "value": city} for city in df["City"]],
            value=df["City"].tolist(),
            multi=True
        ),
        html.Label("Select minimum GDP per capita:"),
        dcc.Slider(
            id="gdp-slider",
            min=df["GDP per capita (2020)"].min(),
            max=df["GDP per capita (2020)"].max(),
            step=10000,
            value=df["GDP per capita (2020)"].min(),
            marks={str(i*1000): str(i*1000) for i in range(1, 11)}
        ),
        html.Div(id="folium-map"),
    ]),
])

# Define the callback function for the folium map
@app.callback(
    Output("folium-map", "children"),
    Input("city-dropdown", "value"),
    Input("gdp-slider", "value")
)
def update_map(selected_cities, min_gdp):
    filtered_df = df[(df["City"].isin(selected_cities)) & (df["GDP per capita (2020)"] >= min_gdp)]
    m = folium.Map(
        location=[df["Latitude"].mean(), df["Longitude"].mean()],
        zoom_start=2,
        tiles="Stamen Terrain"
    )
    marker_cluster = MarkerCluster().add_to(m)
    for index, row in filtered_df.iterrows():
        folium.Marker(
            location=[row["Latitude"], row["Longitude"]],
            popup=f"{row['City']} - GDP per capita: {row['GDP per capita (2020)']}",
        ).add_to(marker_cluster)
    return folium.Html( m._repr_html_(), script=True )

# Run app
if __name__ == '__main__':
    #app.run_server(debug=True)
    # Run app and display result inline in the notebook
    app.run_server(mode='inline')
    #app.run_server(mode='inline',host="0.0.0.0",port=1005)  
```
